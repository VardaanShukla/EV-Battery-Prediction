```python
drop_columns = ['junction_post_tempr(℃)','Total Dchg Cap.( Ngy.)','Total Chg Cap.(Ngy.)','Relay Status']
df.drop(columns = drop_columns,inplace = True)
```

```python
df['Device_Time'] = df['Device_Time'].astype('datetime64[ns]')
df['hour'] = df.Device_Time.dt.hour
df['month'] = df.Device_Time.dt.month
df['day'] = df.Device_Time.dt.day
```

```python
drop_columns = ['Device_Time']
df.drop(columns = drop_columns,inplace = True)
```

```python
df.head()
```

```python
column_to_move = 'Sys_State'
new_position = 65
column = df.pop(column_to_move)
df.insert(new_position, column.name, column)
```

```python
X = df.iloc[:,-1:]
Y = df.iloc[:,:-1]
print(Y)
```

```python
X=pd.get_dummies(X, drop_first=True)
```

```python
df['Sys_State'] = df['Sys_State'].replace({'Charging': 0, 'Discharging': 1})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state = 12)
x_train.shape
```

```python
#standard Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(x_train)
X_test=ss.transform(x_test)
```

```python

X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
```

```python
from sklearn.ensemble import RandomForestRegressor
# Create and train the RandomForestRegressor
RFR = RandomForestRegressor()
RFR.fit(X_train, y_train)
```

```python
Y_pred=RFR.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
# Assuming Y_pred contains the predictions made by your regression model
mae = mean_absolute_error(y_test, Y_pred)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Initialize different regression models
linear_reg = LinearRegression()
decision_tree_reg = DecisionTreeRegressor()
random_forest_reg = RandomForestRegressor()

# Train the models
linear_reg.fit(X_train, y_train)
decision_tree_reg.fit(X_train, y_train)
random_forest_reg.fit(X_train, y_train)

# Make predictions on the test data
y_pred_linear_reg = linear_reg.predict(X_test)
y_pred_decision_tree_reg = decision_tree_reg.predict(X_test)
y_pred_random_forest_reg = random_forest_reg.predict(X_test)

# Evaluate the models using Mean Absolute Error (MAE)
mae_linear_reg = mean_absolute_error(y_test, y_pred_linear_reg)
mae_decision_tree_reg = mean_absolute_error(y_test, y_pred_decision_tree_reg)
mae_random_forest_reg = mean_absolute_error(y_test, y_pred_random_forest_reg)

# Print the MAE for each model
print("Linear Regression Mean Absolute Error:", mae_linear_reg)
print("Decision Tree Regression Mean Absolute Error:", mae_decision_tree_reg)
print("Random Forest Regression Mean Absolute Error:", mae_random_forest_reg)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)  # Adjust the degree as needed
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Train a linear regression model
linear_reg_poly = LinearRegression()
linear_reg_poly.fit(X_train_poly, y_train)

# Make predictions on the test data
y_pred_poly = linear_reg_poly.predict(X_test_poly)

# Evaluate the model using Mean Absolute Error (MAE)
mae_poly = mean_absolute_error(y_test, y_pred_poly)
print("Polynomial Regression Mean Absolute Error:", mae_poly)
```

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense

# Reshape input data to have the correct shape
X_train_reshaped = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])  # Assuming X_train has shape (11904, 1)
X_test_reshaped = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])  # Assuming X_test has the same shape as X_train

# Define the RNN model
model = Sequential([
    SimpleRNN(64, input_shape=(1, X_train.shape[1])),  # Adjust input shape to match the shape of your input data
    Dense(1)  # Output layer with one neuron for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')  # Assuming regression task, so using mean squared error as loss

# Train the model
model.fit(X_train_reshaped, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
loss = model.evaluate(X_test_reshaped, y_test)
print(f'Test Loss: {loss}')
```
